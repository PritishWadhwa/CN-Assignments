```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
plt.style.use('dark_background')
```

# Cwnd vs time

```python
x = []
y = []
with open('tcp-example.cwnd', 'r') as file:
    lines = file.readlines()
    for line in lines:
        x.append(float(line.split()[0]))
        y.append(float(line.split()[1]))
# plt.plot(x, y, 'r', label='cwnd')
plt.scatter(x, y, label='cwnd', color='r', marker='.', s=20)
plt.xlabel('Time')
plt.ylabel('Cwnd')
plt.title('TCP Cwnd')
plt.legend()
plt.show()
```

# Queuing Delay vs Time

```python
startTime = []
endTime = []
with open('tcp-example.tr', 'r') as file:
    lines = file.readlines()
    for line in lines:
        if line[0] == '+':
            timeStart = line.find('+ ') + 2
            timeEnd = line.find(' ', timeStart)
            time = line[timeStart:timeEnd]
            startTime.append(float(time))
        elif line[0] == '-':
            timeStart = line.find('- ') + 2
            timeEnd = line.find(' ', timeStart)
            time = line[timeStart:timeEnd]
            endTime.append(float(time))
qtime = np.array(endTime) - np.array(startTime)
plt.plot(endTime, qtime, 'r', label='queing delay')
plt.xlabel('Time')
plt.ylabel('Queing Delay')
plt.title('TCP Queing Delay')
plt.legend()
plt.show()
```
